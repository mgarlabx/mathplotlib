```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import numpy as np
```

```python
# Dados dos exemplos
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
z = [9, 7, 1, 5]
w = [900, 700, 100, 500]
```

# Figure, Axes & Plots

```python
# Forma completa (1 figure, 1 axes, 1 plot)
fig = plt.figure() # cria uma figura (figure)
ax = fig.add_subplot() # cria um gráfico (axes)
ax.plot(x, y) # cria uma linha (plot)
```

```python
# Forma simplificada (1 figure, 1 axes, 1 plot)
fig, ax = plt.subplots() # cria a figura e o axes
ax.plot(x, y) # cria uma linha (plot)
```

```python
# Forma simplificada (1 figure, 1 axes, 2 plots)
fig, ax = plt.subplots() # cria a figura e o axes
ax.plot(x, y) # cria uma linha (plot)
ax.plot(x, z) # cria outra linha (plot)
```

```python
# Forma simplificada (1 figure, 2 axes, 1 plot cada)
fig, ax = plt.subplots(2, 1) # cria a figura e os axes - 2 linhas e 1 coluna
ax[0].plot(x, y) # cria uma linha (plot)
ax[1].plot(x, z) # cria outra linha (plot)
```

```python
# Forma simplificada (1 figure, 2 axes, 1 plot cada)
fig, ax = plt.subplots(1, 2) # cria a figura e os axes - 1 linha e 2 colunas
ax[0].plot(x, y) # cria uma linha (plot)
ax[1].plot(x, z) # cria outra linha (plot)
```

# Barras

```python
fig, ax = plt.subplots()
ax.bar(x, y)
```

```python
x = np.array(x)
fig, ax = plt.subplots()
ax.bar(x-0.2, y, width=0.4)
ax.bar(x+0.2, z, width=0.4)
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)
ax.bar(x, z, bottom=y)
```

```python
fig, ax = plt.subplots()
ax.barh(x, y)
```

```python
x = np.array(x)
fig, ax = plt.subplots()
ax.barh(x-0.2, y, height=0.4)
ax.barh(x+0.2, z, height=0.4)
```

```python
fig, ax = plt.subplots()
ax.barh(x, y)
ax.barh(x, z, left=y)
```

# Formatações
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

## Cores

```python
# https://matplotlib.org/stable/gallery/color/named_colors.html
fig, ax = plt.subplots()
ax.plot(x, y, color='r') 
```

## Estilo

```python
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
fig, ax = plt.subplots()
ax.plot(x, y, linestyle='--') 
```

## Largura

```python
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=5)
```

## Marcador

```python
fig, ax = plt.subplots()
ax.plot(x, y, marker='o')
```

## Combinando em um só comando

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'o--r') # [marker][line][color]
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, '*:b') # [marker][line][color]
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, '*:b', ms=20, mec='b', mfc='y') # ms = tamanho, mec = cor da borda, mfc = cor do recheio
```

## Área sob o gráfico

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x, y, color='cyan', alpha=0.2) # alpha = transparência
```

```python
fig, ax = plt.subplots()
ax.fill_between(x, y, color='cyan', alpha=0.2) # alpha = transparência
```

# Eixos

```python
# Espaçamento horizontal
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.xaxis.set_major_locator(plticker.MultipleLocator(base=1))
```

```python
# Espaçamento vertical
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.yaxis.set_major_locator(plticker.MultipleLocator(base=2))
```

```python
# Formatação dos ticks
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.tick_params(axis='both', which='major', labelsize=50)
```

```python
# limites (faixas) dos eixos
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.set_xlim(2, 4)
ax.set_ylim(0, 10)
```

```python
# Eixo secundário
fig, ax = plt.subplots()
ax.plot(x, y) 
ax2 = ax.twinx()
ax2.plot(x, w, color='g') 
```

# Bordas e grids

```python
# remove a caixa ao redor do gráfico
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.axis('off')
```

```python
# remove ticks
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.set_xticks([])
ax.set_yticks([])
```

```python
# cor do fundo
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.set_facecolor('y')
```

```python
# grid - linhas horizontais
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.yaxis.grid() 
```

```python
# grid - linhas verticais
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.xaxis.grid() 
```

```python
# duplo grid
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.xaxis.grid() 
ax.yaxis.grid() 
```

# Títulos e legendas

```python
# Título
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.set_title('Lorem')
```

```python
# Títulos
fig, ax = plt.subplots(1, 2) 
ax[0].plot(x, y)
ax[1].plot(x, z)
ax[0].set_title('Lorem')
ax[1].set_title('Ipsum')
```

```python
# Legendas
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.set_xlabel('Lorem')
ax.set_ylabel('Ipsum')
```

### Textos avulsos

```python
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.text(3, 2, 'Lorem')
```

```python
fig, ax = plt.subplots()
ax.plot(x, y) 
ax.text(min(x), max(y), 'Lorem')
```
